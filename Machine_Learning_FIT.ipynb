## Problemas

* Locomoção afetada pelos terrenos
* Escassez de dados coorelacionados ao desastre
* Falta de infraestrutura gera a demora no atendimento da ocorrência ocasionando uma maior área destruída pelo incêndio

### Causas e efeitos

* Causa: A seca severa
  * Efeito: Liberação de gases de efeito estufa na atmosfera impactando em mudanças climáticas

* Causa: Efeito de borda
  * Efeito: Processos de degradação tem seu perímetro e exposição da mata aumentados

* Causa: Crimes ambientais
  * Efeito: Acidentes incendiários ou propositais nos locais

## Área de Aplicação

* Identificação e alerta em pontos de ignição.

## Objetivos de Negócio
  * Atender as ocorrências com rapidez
  * Rápido processamento de informação sobre os incêndios
  * Anular e/ou remediar os impacto no meio ambiente

  * Critérios de sucesso:
    * Restauração rápida ou considerável da fauna e flora
    * Conter a emissão de CO2
    * Minimizar perdas na biodiversidade

## Objetivos de Machine Learning
  * Usar os dados para que se possa prevenir tais catástrofes
  * Localizar e reportar as ocorrências em tempo real
  * Apresentar e analisar lugares com probabilidade de incêndio

  * Critérios de sucesso
    * Redução percentual consequente dos desastres
    * Encontrar o melhor caminho até a origem do incêndio
    * Diminuir o tempo de resposta dos combatentes

```python
#@title Cronograma
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.DataFrame([
    dict(tarefa="Entrega", inicio=14, fim=15),
    dict(tarefa="Avaliação do modelo", inicio=12, fim=14),
    dict(tarefa="Modelagem", inicio=9, fim=12),
    dict(tarefa="Preparação dos dados", inicio=4, fim=9),
    dict(tarefa="Entendimento dos dados", inicio=2, fim=4),
    dict(tarefa="Entendimento do negócio", inicio=0, fim=2),
])

df['semanas'] = df.fim - df.inicio

fig, ax = plt.subplots(1, figsize=(16,6))
ax.barh(df.tarefa, df.semanas, left=df.inicio)
ax.set_axisbelow(True)
ax.xaxis.grid(color='gray', linestyle='dashed', alpha=0.2, which='both')

xticks = np.arange(0, df.fim.max()+1, 1)
xticks_minor = np.arange(0, df.fim.max()+1, 1)
ax.set_xticks(xticks)
ax.set_xticks(xticks_minor, minor=True)

plt.show()
```
