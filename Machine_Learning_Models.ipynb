```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Testes para modelos
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
# Leitura das bases
df_train = pd.read_csv('/content/drive/MyDrive/IA/Machine Learning/_focos_treino.csv')
df_test = pd.read_csv('/content/drive/MyDrive/IA/Machine Learning/_focos_teste.csv')

# Objetivo h√° se atingir
y_train = df_train['alvo']
y_test = df_test['alvo']

# Conjunto de treinamento
X_train = df_train.drop('alvo', axis=1)
X_test = df_test.drop('alvo', axis=1)
```

## Classifica√ß√£o

```python
# √Årvore de decis√£o (DT)
from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier(criterion='gini', max_depth=10)

# Treinamento para tomada de decis√µes
dt_classifier = dt_classifier.fit(X_train, y_train)

# Criar predi√ß√£o na base de testes
dt_predict = dt_classifier.predict(X_test)

# Verificar acur√°cia
dt_acc = accuracy_score(dt_predict, y_test)
print("DT Acur√°cia", dt_acc)
```

```python
# K-vizinhos mais pr√≥ximos (KNN)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=10)

# Treinamento para tomada de decis√µes
knn = knn.fit(X_train, y_train)

# Criar predi√ß√£o na base de testes
knn_predict = knn.predict(X_test)

# Verificar acur√°cia
knn_acc = accuracy_score(knn_predict, y_test)
print("KNN Acur√°cia", knn_acc)
```

## Regress√£o

```python
# Regress√£o Linear (LR)
from sklearn.linear_model import LinearRegression

lr  = LinearRegression().fit(X_train, y_train)

# Predi√ß√£o para a base teste
lr_predict = lr.predict(X_test)

# Avalia√ß√£o do erro
print("R^2 :", r2_score(y_test, lr_predict))
print("Erro m√©dio absoluto :", mean_absolute_error(y_test, lr_predict))
print("Raiz do Erro Quadr√°tico M√©dio:", np.sqrt(mean_squared_error(y_test, lr_predict)))
```

## Multi Layer Perceptrons

```python
# Rede neural de classifica√ß√£o
from sklearn.neural_network import MLPClassifier

# https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPClassifier.html
clf = MLPClassifier( # Ajustar os hiperparametros
    hidden_layer_sizes = (10, 12, 30, 60, 30, 5, 50, 6, 25),
)
clf = clf.fit(X_train, y_train)

# Predi√ß√£o da rede com a base de teste
clf_predict = clf.predict(X_test)

# Resultado obtido com a predi√ß√£o
clf_acc = accuracy_score(clf_predict, y_test)
print("CLF Acur√°cia", clf_acc)
```

## Modelo Selecionado

```python
'''
  Este algoritmo seleciona √°rvores de decis√£o
  aleat√≥rias em v√°rias subamostras e usa a m√©dia para melhorar
  a precis√£o preditiva e controlar o sobreajuste (overfitting)
'''

# Floresta Aleat√≥ria
from sklearn.ensemble import RandomForestClassifier

# Treinamento para tomada de decis√µes
model = RandomForestClassifier(
    criterion = 'entropy',
    max_depth = 70,
    min_samples_split = 5,
    min_samples_leaf = 5,
).fit(X_train, y_train)

# Criar predi√ß√£o na base de testes
model_predict = model.predict(X_test)

# Verificar acur√°cia
model_acc = accuracy_score(model_predict, y_test)
print("Modelo Acur√°cia:", model_acc)
```

```python
print('Acur√°cia de Treinamento :', accuracy_score(y_train, model.predict(X_train)))
print('Acur√°cia de Teste :', accuracy_score(y_test, model.predict(X_test)))
```
